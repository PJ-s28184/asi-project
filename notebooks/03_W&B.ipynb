```python
import wandb
import pandas as pd

api = wandb.Api()

PROJECT = "asi-project"

runs = api.runs(PROJECT)

records = []
for run in runs:
    metrics = run.summary
    config = run.config
    records.append({
        "run_id": run.id,
        "name": run.name,
        "mae": metrics.get("mae"),
        "mape": metrics.get("mape"),
        "rmse": metrics.get("rmse"),
        "mse": metrics.get("mse"),
        "r2": metrics.get("r2"),
        "_runtime": metrics.get("_runtime"),
    })

df = pd.DataFrame(records)
df.sort_values("mae")
```

```python
api = wandb.Api()
artifact = api.artifact("asi-project/run-smi3qwbd-history:v0")
artifact.aliases.append("production")
artifact.save()
```
